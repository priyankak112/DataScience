# IQR InterQuartile Range-Lesser outlier,UpperOutlier calculation


```python
descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25","Q2:50","Q3:75","99th","Q4:100","IQR","1.5rule","LesserBound","UpperBound","min","max"],columns=quan)
for col in quan:
        descriptive.loc["Mean",col]=dataset[col].mean()
        descriptive.loc["Median",col]=dataset[col].median()
        descriptive.loc["Mode",col]=dataset[col].mode()[0]
        descriptive.loc["Q1:25",col]=dataset.describe()[col]["25%"]
        descriptive.loc["Q2:50",col]=dataset.describe()[col]["50%"]
        descriptive.loc["Q3:75",col]=dataset.describe()[col]["75%"]
        descriptive.loc["99th",col]=np.percentile(dataset[col],99)
        descriptive.loc["Q4:100",col]=dataset.describe()[col]["max"]
        descriptive.loc["IQR",col]=descriptive.loc["Q3:75",col]-descriptive.loc["Q1:25",col]
        descriptive.loc["1.5rule",col]=1.5*descriptive[col]["IQR"]
        descriptive.loc["LesserBound",col]=descriptive[col]["Q1:25"]-descriptive[col]["1.5rule"]
        descriptive.loc["UpperBound",col]=descriptive[col]["Q3:75"]+descriptive[col]["1.5rule"]
        descriptive.loc["min",col]=dataset[col].min()
        descriptive.loc["max",col]=dataset[col].max()
```

# Finding Outliers-"<" lesserbound and ">" greaterbound

```python
lesser_outlier=[]
upper_outlier=[]
for col in quan:
    if (descriptive[col]["min"]<descriptive[col]["LesserBound"]):
        lesser_outlier.append(col)
    if(descriptive[col]["max"]>descriptive[col]["UpperBound"]):
        upper_outlier.append(col)
```
