```python
def pdf_calc(data,first,end):
    import seaborn as sns
    from matplotlib import pyplot
    from scipy.stats import norm
    import numpy as np
    ax=sns.displot(data,kde=True,line_kws={'color':'red'},color='purple')
    pyplot.axvline(first,color='green')
    pyplot.axvline(end,color='green')
    mn=data.mean()
    sd=data.std()
    dist=norm(mn,sd)
    probabilities=list(map(lambda v:dist.pdf(v),range(first,end)))
    total_pdf=sum(probabilities)
    print(f"Total PDF from {first} to {end} is: {total_pdf}")
    return total_pdf
```

```python
pdf_calc(df["ssc_p"],40,65)
```

```python
#z score calculate
def z_score(data):
    import seaborn as sns
    import numpy as np
    mean1=data.mean()
    sd=data.std()
    val=np.array(data)
    z_scr=(val-mean1)/sd
    sns.displot(z_scr,kde=True)
    print("avg zscore::",z_scr.mean())
    
```

```python
z_score(df["ssc_p"])
```
