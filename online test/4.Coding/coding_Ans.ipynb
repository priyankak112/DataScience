```python
#A company wants to analyze customer satisfaction survey data. How can Data Science concepts help to clean and analyze this data?
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
data={'CustomerID': [1, 2, 3, 4, 5, 5, 6, 7, 8, 9],
'Satisfaction': [8, 7, np.nan, 9, 6, 6, 10, 4, 5, 7]}
df=pd.DataFrame(data)
df.fillna(0,inplace=True)
df
mean=df['Satisfaction'].mean()
median=df['Satisfaction'].median()
mode=stats.mode(df['Satisfaction']).mode
std=df['Satisfaction'].std()
print("Mean:",mean)
print("median:",median)
print("mode:",mode)
print("std:",std)
df.describe()
```

```python
# A retail store has recorded monthly sales data for its products. How can Data Science concepts be applied to understand sales performance?
data={'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
 'Sales': [1500, 1700, 1600, 1800, 2200, 2100]}
df=pd.DataFrame(data)
mean=df['Sales'].mean()
median=df['Sales'].median()
mode=stats.mode(df['Sales']).mode
std=df['Sales'].std()
print("Mean:",mean)
print("median:",median)
print("mode:",mode)
print("std:",std)
df.describe()
```

```python
#A hospital conducted a study to check the effectiveness of a new medical test. How can Data Science concepts be used to validate its results?

group_a = [110, 115, 108, 102, 107]
group_b = [120, 125, 130, 122, 128]
ttest,pval=stats.ttest_ind(group_a,group_b)
print("ttest",ttest)
print(f"pval,{pval:.4f}")
threshold=0.05
if pval<threshold:
    print("Reject null hypothesis-The test is effective")
else:
    print("Accept null hypothesis-not effective")
```

```python
# A company wants to evaluate employee performance using various metrics. How can Data Science concepts be applied?

data={'Department': ['HR', 'IT', 'Sales', 'IT', 'HR', 'Sales', np.nan, 'HR', 'Sales', 'IT'],
    'Experience (Years)': [1.0, 3.0, 5.0, np.nan, 10.0, 2.0, 8.0, 4.0, np.nan, 9.0],
    'Projects Completed': [2.0, 5.0, np.nan, 10.0, 12.0, 3.0, 9.0, np.nan, 14.0, 11.0],
    'Performance Rating': [3.0, np.nan, 7.0, 8.0, 9.0, 4.0, 8.0, 7.0, 10.0, np.nan]}
df=pd.DataFrame(data)
#data cleaning
df.fillna(0,inplace=True)
df
```

```python
#2)	Descriptive Statistics
df.describe()
```

```python
#3)	Correlation Analysis (Using heatmap)
#df.columns
corr_matrix=df[['Projects Completed','Performance Rating']].corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix,annot=True,cmap='Purples')
plt.title("Heatmap")
plt.xlabel("correlation")
plt.show()
```

```python
df[['Experience (Years)','Projects Completed','Performance Rating']].hist(bins=5,figsize=(5,5),color='orange')
plt.title("Histogram")
plt.show()
```
